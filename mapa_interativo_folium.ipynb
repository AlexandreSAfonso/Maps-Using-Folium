```python
import folium
from folium import plugins
```

```python
sete_maravilhas_nomes = [
    "Grande Piramide Gizé", 
    "Jardins Suspensos da Babilônia", 
    "Templo de Artemis", 
    "Estatua de Zeus em Olímpia", 
    "Mausoléo de Halicarnasso", 
    "Colosso de Rodes", 
    "Farol de Alexandria"
    ]

sete_maravilhas_coordenadas = [
    [29.9792, 31.1344], 
    [32.5355, 44.4274], 
    [37.9497, 27.3639], 
    [37.6378, 21.6299], 
    [37.0378, 27.424], 
    [36.451, 28.2277], 
    [31.2139, 29.8856]
    ]
```

```python
sete_maravilhas_mapa = folium.Map(location=[34.4469, 30.4930], zoom_start=5, control_scale=True, tiles="cartodbpositron")

# sete_maravilhas_mapa = folium.Map(location=sete_maravilhas_coordenadas[0], zoom_start=7, control_scale=True)
# folium.TileLayer("cartodbpositron").add_to(sete_maravilhas_mapa)

# folium.TileLayer(
#     tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Street_Map/MapServer/tile/{z}/{y}/{x}',
#     attr='Tiles &copy; Esri &mdash; Source: Esri, DeLorme, NAVTEQ, USGS, Intermap, iPC, NRCAN, Esri Japan, METI, Esri China (Hong Kong), Esri (Thailand), TomTom, 2012',
#     name='Esri.WorldStreetMap'
#     ).add_to(sete_maravilhas_mapa)

#folium.LayerControl().add_to(sete_maravilhas_mapa)

# add landmarks
for i in range (len(sete_maravilhas_coordenadas)):
    coodenada = sete_maravilhas_coordenadas[i]
    maravilha = sete_maravilhas_nomes[i]
    folium.Marker(coodenada, 
                  popup=maravilha, 
                  icon=folium.Icon(
                      icon='glyphicon glyphicon-camera',
                      color='black',
                      icon_color='white',
                      prefix='glyphicon'
                      )
                  ).add_to(sete_maravilhas_mapa)

sete_maravilhas_mapa.add_child(folium.LatLngPopup())

sete_maravilhas_mapa.save('sete_maravilhas_mapa.html')

sete_maravilhas_mapa
```
